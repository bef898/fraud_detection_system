Bivariate Analysis

I used boxplots to see how purchase_value varies between fraud (class = 1) and non-fraud (class = 0).

```python
sns.boxplot(x='class', y='purchase_value', data=fraud_data)
plt.title('Purchase Value by Fraud Class')
plt.show()
```

```python
sns.countplot(x='browser', hue='class', data=fraud_data)
plt.title('Fraud Cases by Browser')
plt.show()
```

```python
sns.violinplot(x='class', y='age', data=fraud_data)
plt.title('Age Distribution by Fraud Class')
plt.show()
```
